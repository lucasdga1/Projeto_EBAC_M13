Os campos do nosso dataframe são:

**Title:** Nome do produto.


**Marca:** A marca do produto.


**Preco_Normal:** O preço em que o produto costuma ser vendido quando não há desconto.


**Preco_Desconto:** O preço vendido após o desconto ser aplicado.


**Preco_Anterior:** Preço em que era comercializado o produto antes do desconto aplicado.


**Desconto:** Total de desconto aplicado.






As colunas que aparecem com valores 0 são para os produtos onde não tivemos descontos aplicados.


As categorias estão em espanhol!

# 1 - Traga a média e a mediana dos preços - coluna Preco_Normal - por categoria de produto.
# Identifique as categorias que parecem ter um valor de média abaixo ou acima da mediana.

```python
#Seu código aqui para a média
df.groupby('Categoria')['Preco_Normal'].mean().reset_index().sort_values(by='Categoria', ascending=False)
```

```python
#Seu código aqui para a mediana
df.groupby('Categoria')['Preco_Normal'].median().reset_index().sort_values(by='Categoria', ascending=False)
```

Digite aqui as categorias:
* As verduras apresentam uma diferença de quase 200 a mais da média em relação a mediana;
* Os lacteos apresentam uma diferença positiva de quase 1400;
* instantaneos-y-sopas apresentam uma diferença positiva de aproximadamente 300;
* frutas apresentam uma diferença positiva de quase 600;
* congelados apresentam uma diferença positiva de quase 600 também;
* As comidas preparadas apresentam uma diferença negativa de aproximadamente 200;
* Beleza e cuidado pessoal apresenta uma diferença positiva de pouco mais de 200.

# 2 - Traga o desvio padrão por categoria de produto.
# Qual o comportamento da média e mediana nas categorias com maior desvio?

```python
#Seu código aqui
df.groupby('Categoria')['Preco_Normal'].std().reset_index().sort_values(by='Categoria', ascending=False)
```

Digite nesse campo o comportamento que você identificou.
O comportamento evidenciado pela análise é que o desvio padrão tende a revelar as categorias com maior diferença entre a mediana e a média, como evidenciado no caso dos lacteos. Porém nos casos de congelados e frutas o desvio padrão se assemelhou à média.

# 3 - Plot um boxplot da distribuição do Preco_Normal para a categoria que você identificou que tem o maior desvio padrão. Como é a distribuição desses dados segundo o boxplot? Você identifica muitos outliers?

Dica: Para trazer apenas os dados da categoria que você deseja você pode usar o df.loc[df['Categoria'] == 'CATEGORIA ESCOLHIDA'

```python
#Seu código aqui
df_lacteos = df_lacteos = df[df['Categoria'].str.lower() == 'lacteos'].copy()
fig = px.box(df_lacteos, x="Categoria", y="Preco_Normal", points='all')
fig.show()
```

A distribuição dos peços na categoria de lacteos se concentra abaixo da mediana, mas apresenta muitos outliers.

# 4 - Plote um gráfico de barras onde temos a média de descontos por categoria.

```python
# Seu código aqui
df_mean = df.groupby('Categoria')['Desconto'].mean().reset_index().sort_values(by='Categoria', ascending=False)
fig = px.bar(df_mean, x= 'Categoria', y='Desconto', color='Categoria', text_auto=True, template='simple_white')
fig.update_yaxes(showticklabels=False, ticks="")
fig.show()
```

# 5 - Plote um gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.

```python
# Seu código aqui
df_map = df.groupby(['Categoria', 'Marca'])['Desconto'].mean().reset_index().sort_values(by='Categoria')
fig = px.treemap(df_map, path =['Categoria', 'Marca'], values = 'Desconto')
fig.show()
```
